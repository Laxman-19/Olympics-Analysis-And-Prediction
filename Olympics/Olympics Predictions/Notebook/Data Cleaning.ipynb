```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../Data/processed_data.csv')
```

```python
df.head()
```

## Handling Nulls

```python
df.isna().sum()
```

```python
allnull = df[df['Age'].isna() & df['Height'].isna() & df['Weight'].isna() & df['Medal'].isna()]
allnull.head(3)
```

```python
df = df[~df.index.isin(allnull.index)]
df.head(3)
```

```python
age_mean = df['Age'].mean()
height_mean = df['Height'].mean()
weight_mean = df['Weight'].mean()

df['Age'].fillna(age_mean, inplace=True)
df['Height'].fillna(height_mean, inplace=True)
df['Weight'].fillna(weight_mean, inplace=True)

df['Medal'].fillna("None", inplace=True)
```

```python
df.isna().sum()
```

## Exporting the clean data

```python
df['Age'] = df['Age'].astype('int')
df['Height'] = df['Height'].astype('int')
df['Weight'] = df['Weight'].astype('int')
df['Year'] = df['Year'].astype('int')
```

```python
df.head(3)
```

```python
df.to_csv('../Data/clean_athlete.csv', index=False)
```
